```python
import pandas as pd
import py_entitymatching as em
```

```python
data1 = em.read_csv_metadata('UWM_Final.csv', delimiter = '\t')
data1.columns = ['ID', 'Title','Author','Publication','Format','ISBN','Series', 'Physical Details']
data1.iloc[:,0] = data1.index.values.astype(int)
data1['ID'] = 'a' + data1['ID'].astype(str)
data1['Title'] = '"' + data1['Title'].astype(str) + '"'
data1['Author'] = '"' + data1['Author'].astype(str) + '"'
data1['Publication'] = '"' + data1['Publication'].astype(str) + '"'
data1['Format'] = '"' + data1['Format'].astype(str) + '"'
data1['Series'] = '"' + data1['Series'].astype(str) + '"'
data1['Physical Details'] = '"' + data1['Physical Details'].astype(str) + '"'
```

```python
data2 = em.read_csv_metadata('UIUC_Final.csv', delimiter = '\t')
data2.columns = ['ID', 'Title','Author','Publication','Format','ISBN','Series', 'Physical Details']
data2.iloc[:,0] = data2.index.values.astype(int)
data2['ID'] = 'b' + data2['ID'].astype(str)
data2['Title'] = '"' + data2['Title'].astype(str) + '"'
data2['Author'] = '"' + data2['Author'].astype(str) + '"'
data2['Publication'] = '"' + data2['Publication'].astype(str) + '"'
data2['Format'] = '"' + data2['Format'].astype(str) + '"'
data2['Series'] = '"' + data2['Series'].astype(str) + '"'
data2['Physical Details'] = '"' + data2['Physical Details'].astype(str) + '"'
```

```python
import re
def ISBN_trans(x):
    ISBN = re.compile('^9\d+')
    if ISBN.search(x):
        return (int)(ISBN.search(x).group(0))
    else:
        return None
```

```python
data1 = data1[~data1['ISBN'].apply(ISBN_trans).isnull()]
data1['ISBN'] = data1['ISBN'].apply(ISBN_trans)
```

```python
data2 = data2[~data2['ISBN'].apply(ISBN_trans).isnull()]
data2['ISBN'] = data2['ISBN'].apply(ISBN_trans)
```

```python
data1.to_csv('TableA.csv', sep = ',')
data2.to_csv('TableB.csv', sep = ',')
```
